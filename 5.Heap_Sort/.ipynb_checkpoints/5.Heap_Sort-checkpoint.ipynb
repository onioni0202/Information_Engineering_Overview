```python
from collections import deque
import networkx as nx
from IPython.display import SVG, display
import pydot
import os
```

```python
#N: 入力数
#A: 配列

N = 6
A = [4, 2, 5, 3, 1, 6]
```

```python
G = nx.Graph()
graph = nx.nx_agraph.to_agraph(G)

def print_graph(A, a1=1, a2=1, color="white"):
    global graph
    G.clear()
    G.add_nodes_from(A)
    for i in range(N//2):
        child1 = i*2 + 1
        child2 = i*2 + 2
        if child1 < len(A):
            G.add_edge(A[i], A[child1])
        if child2 < len(A):
            G.add_edge(A[i], A[child2])
    graph = nx.nx_agraph.to_agraph(G)
    graph.node_attr["shape"] = "circle"
    graph.node_attr["style"] = "filled"
    for i in range(1, N+1):
        n = graph.get_node(i)
        n.attr['fillcolor'] = "white"
    n = graph.get_node(a1)
    n.attr['fillcolor'] = color
    n = graph.get_node(a2)
    n.attr['fillcolor'] = color
    svg = SVG(graph.draw(prog='dot', format='svg'))
    display(svg)

picnum = 0
def save_graph(name=""):
    global picnum
    if name == "":
        name = str(picnum)
        picnum += 1
    graph.write("svg/"+name+".dot")
    (f,) = pydot.graph_from_dot_file("svg/"+name+".dot")
    f.write_svg("svg/"+name+'.svg')
    os.remove("svg/"+name+".dot")
```

```python
print_graph(A)
#save_graph("default")

def Heapify(A, i, n):
    child1 = i*2 + 1
    child2 = i*2 + 2
    if child1 >= n:
        return
    if child2 < n and A[child2] > A[child1]:
        child1 = child2
    if A[child1] <= A[i]:
        return
    A[i], A[child1] = A[child1], A[i]
    print_graph(A, A[i], A[child1], "red")
    #save_graph()
    Heapify(A, child1, n)


def HeapSort(A):
    for i in range(N//2)[::-1]: # 最後尾から順番にヒープ
        Heapify(A, i, N)

    for i in range(N)[::-1]:
        A[0], A[i] = A[i], A[0]  # 最大値であるAの先頭と、大小が確定していない最後尾を入れ替え
        print_graph(A, A[0], A[i], "yellow")
        #save_graph()
        Heapify(A, 0, i) #先頭が最大ではなくなったのでヒープ

HeapSort(A)
```
