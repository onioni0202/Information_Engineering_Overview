```python
from collections import defaultdict, deque
import graphviz
import os
```

```python
# N: 男性と女性の人数は等しくN人
# m: 女性1~Nの中で好きな人の順番の情報をN個入力
# l: 男性1~Nの中で好きな人の順番の情報をN個入力

N = 5
M = [[3, 5, 1, 4, 2],
     [1, 2, 3, 4, 5],
     [1, 4, 3, 5, 2],
     [2, 5, 4, 1, 3],
     [4, 1, 5, 2, 3]]
L = [[4, 1, 3, 2, 5],
     [5, 1, 2, 3, 4],
     [2, 3, 1, 5, 4],
     [3, 4, 2, 5, 1],
     [3, 1, 5, 2, 4]]

for i in range(N):
    M[i] = deque(M[i])
    L[i] = deque(L[i])
```

```python
def print_graph():
    display(G)
    print()
    print()
    print()
    
def add_line_and_color(m, l):
    G.edge(str(m)+"M", str(l)+"L", color="#000000")
    G.node(str(m)+"M"+str(l)+"L", fillcolor="yellow")
    G.node(str(l)+"L"+str(m)+"M", fillcolor="yellow")
    
def remove_line_and_color(m, l):
    G.edge(str(m)+"M", str(l)+"L", color="#00000000")
    G.node(str(m)+"M"+str(l)+"L", fillcolor="#87cefa")
    G.node(str(l)+"L"+str(m)+"M", fillcolor="#ff69b4")
    
picnum = 0
def save_graph(name=""):
    global picnum
    if name == "":
        name = str(picnum)
        picnum += 1
    G.render("svg/"+name)
    os.remove("svg/"+name)
```

```python
G = graphviz.Graph(engine='neato', format='svg', strict=True)

for i in range(N):
    for g in range(N):
        G.node(str(i+1)+"M"+str(M[i][g])+"L",
               label=str(M[i][g]), 
               pos=str(0.6*g)+","+str(0.6*(N-i))+"!", 
               style='filled',
               fillcolor='#87cefa',
               shape="circle", 
               **{'width':"0.5", 'height':"0.5"})
        
for i in range(N):
    G.node(str(i+1)+"M",
           pos=str(0.6*(N-1+0.8))+","+str(0.6*(N-i))+"!",
           style='filled',
           fillcolor='black',
           shape="circle",
           fontsize="0",
           fixedsize = "true",
           **{'width':"0.05", 'height':"0.05"})
        
for i in range(N):
    for g in range(N):
        G.node(str(i+1)+"L"+str(L[i][g])+"M",
               label=str(L[i][g]), 
               pos=str(4+0.6*g)+","+str(0.6*(N-i))+"!", 
               style='filled',
               fillcolor='#ff69b4',
               shape="circle", 
               **{'width':"0.5", 'height':"0.5"})

for i in range(N):
    G.node(str(i+1)+"L",
           pos=str(4-0.6*0.8)+","+str(0.6*(N-i))+"!",
           style='filled',
           fillcolor='black',
           shape="circle",
           fontsize="0",
           fixedsize = "true",
           **{'width':"0.05", 'height':"0.05"})
    
for i in range(N):
    for g in range(N):
        G.edge(str(i+1)+"M", str(g+1)+"L", color="#00000000")

display(G)
#save_graph("default")
```

```python
Mpair = defaultdict(int) # 0->ペアなし、0以外->パートナーの女性
Lpair = defaultdict(int) # 0->ペアなし、0以外->パートナーの男性
while True:
    for man1 in range(1, N+1):
        if Mpair[man1] == 0: # 男性にペアがいない
            lady = M[man1-1][0] # 今最もペアになりたい対象の女性
            M[man1-1].popleft()
            if Lpair[lady] == 0: # 女性にペアがいない場合
                add_line_and_color(man1, lady)
                Mpair[man1] = lady
                Lpair[lady] = man1
            else: #女性にペアがいる場合
                man2 = Lpair[lady]
                for i in range(N):
                    if L[lady-1][i] == man1: # man1の方が好き
                        Mpair[man1] = lady
                        Mpair[man2] = 0
                        Lpair[lady] = man1
                        add_line_and_color(man1, lady)
                        remove_line_and_color(man2, lady)
                        break
                    if L[lady-1][i] == man2: # man2の方が好き
                        Mpair[man1] = 0
                        Mpair[man2] = lady
                        Lpair[lady] = man2
                        add_line_and_color(man2, lady)
                        remove_line_and_color(man1, lady)
                        break
        print(man1)
        print_graph()
        #save_graph()
    
    # 男性全員にペアがいるならば終了
    end = True
    for man1 in range(1, N+1):
        if Mpair[man1] == 0:
            end = False
    if end:
        break
```

```python
for m,l in Mpair.items():
    print("男性{0}のペアは女性{1}です。".format(m,l))
```
