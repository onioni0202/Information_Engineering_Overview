```python
import heapq
import graphviz
import os
```

```python
# N:ノードの数　W:エッジ(辺)の数
# xyz (x:始点、y:終点、z:重み)

N = 7
W = 9
s = 0 # スタート
xyz = [(0, 1, 1),
       (1, 2, 2),
       (1, 3, 3),
       (2, 5, 10),
       (1, 4, 7),
       (3, 4, 1),
       (3, 6, 4),
       (4, 6, 8),
       (4, 5, 5)]


graph = [[] for i in range(N)]

for i in range(W):
    graph[xyz[i][0]].append((xyz[i][1], xyz[i][2]))
    graph[xyz[i][1]].append((xyz[i][0], xyz[i][2]))
```

```python
def print_graph(used):
    for i, u in enumerate(used):
        if u == True:
            G.node(str(i), style="filled", fillcolor='red')        
    display(G)
        
picnum = 0
def save_graph(name=""):
    global picnum
    if name == "":
        name = str(picnum)
        picnum += 1
    G.render("svg/"+name)
    os.remove("svg/"+name)
```

```python
G = graphviz.Graph(engine='neato', format='svg', strict=True)
positions = ["0,0!", "1,0!", "1,-1!", "1.4,1.2!", "4,-1!", "2,-1!" ,"5,1!"]
for i in range(N):
    G.node(str(i), pos=positions[i], shape='circle')
for i in range(W):
    G.edge(str(xyz[i][0]), str(xyz[i][1]), label=str(xyz[i][2])) 
```

```python
def dijkstra_heap(s):
    dist = [float("inf")] * N
    used = [False] * N  # True:確定済み
    dist[s] = 0
    used[s] = True
    edgelist = []
    for e in graph[s]:
        heapq.heappush(edgelist, e[::-1])
    v = 1
    while v < N:
        minedge = heapq.heappop(edgelist) # minedge[0]->重み, minedge[1]->向かう先
        if used[minedge[1]]:
            continue
        dist[minedge[1]] = minedge[0]
        used[minedge[1]] = True
        v += 1
        for e in graph[minedge[1]]:
            if not used[e[0]]:
                heapq.heappush(edgelist, (e[1] + minedge[0], e[0]))
        print_graph(used)
        #save_graph()
    return dist
```

```python
G.node(str(s), style="filled", fillcolor='red')  
display(G)
#save_graph("default")
dijkstra_heap(0)
```

```python
def print_path(path):
    #初期化
    for i in range(N):
        G.node(str(i), style="filled", fillcolor='white')
    for x, y, z in xyz:
        G.edge(str(x), str(y), color='black', penwidth="1")
    #パスに色をつける
    for p in path:
        G.node(str(p), style="filled", fillcolor='yellow')
    for i in range(len(path)-1):
        G.edge(str(path[i]), str(path[i+1]), color='red', penwidth="3")
    display(G)
```

```python
# 経路復元(実装はダイクストラというよりも幅優先探索に近い)

dist = [float("inf")] * N # 最短距離の値
prev = [-1] * N # 経路を記録

def dijkstra_back(s):
    dist[s] = 0
    que = []
    heapq.heappush(que, (0, s))

    while len(que) != 0:
        p = heapq.heappop(que)
        v = p[1]
        if (dist[v] < p[0]):
            continue
        for e in graph[v]: # e[0]->向かう先、e[1]->重み
            if dist[e[0]] > dist[v] + e[1]:
                dist[e[0]] = dist[v] + e[1]
                prev[e[0]] = v
                heapq.heappush(que, (dist[e[0]], e[0]))
```

```python
def get_path(t):
    path = []
    while t != -1:
        path.append(t)
        t = prev[t]
    path = path[::-1]
    return path
```

```python
dijkstra_back(0)
for i in range(N):
    print_path(get_path(i))
    #save_graph("path"+str(i))
```
