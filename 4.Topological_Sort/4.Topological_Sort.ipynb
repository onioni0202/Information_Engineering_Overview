```python
from collections import defaultdict, deque
import graphviz
import os
```

```python
# N:=ノード数 
# M:=辺数
# xy: エッジの始点と終点

N = 8
M = 8
xy = [(4, 1),
      (3, 1),
      (3, 2),
      (1, 5),
      (4, 6),
      (1, 6),
      (6, 7),
      (7, 8)]

graph = defaultdict(list)
V = defaultdict(int)
for i in range(M):
    graph[xy[i][0]].append(xy[i][1])
    V[xy[i][1]] += 1
```

```python
num = 0
def move_node(v):
    global num
    G.node(str(v), style="filled", fillcolor="yellow", pos=str(num*0.6)+",0!", shape='circle')
    if v%2 == 0:
        for to in graph[v]:
            G.edge(str(v), str(to), headport="nw", tailport="ne")
    else:
        for to in graph[v]:
            G.edge(str(v), str(to), headport="sw", tailport="se")
    num += 1
    
picnum = 0
def save_graph(name=""):
    global picnum
    if name == "":
        name = str(picnum)
        picnum += 1
    G.render("svg/"+name)
    os.remove("svg/"+name)
```

```python
G = graphviz.Digraph(engine='neato', format='svg', strict=True)
G.attr('graph', splines='true') 
positions = ["2,0!", "3,-1!", "5,0.5!", "3.4,1.2!", "5,-0.3!", "4.2,1!" ,"5,-1!","4.2,-1!"]
for i in range(1,N+1):
    G.node(str(i), pos=positions[i-1], shape='circle')
for i in range(M):
    G.edge(str(xy[i][0]), str(xy[i][1]))
display(G)
#save_graph("default")
```

```python
q = deque()
for v in range(1, N):
    if V[v] == 0:
        q.append(v)
        move_node(v)
        display(G)
        save_graph()

ans = []
while q:
    From = q.popleft()
    ans.append(From)
    for to in graph[From]:
        V[to] -= 1
        if V[to] == 0:
            q.append(to)
            move_node(to)
            display(G)
            #save_graph()
print(ans)
```
