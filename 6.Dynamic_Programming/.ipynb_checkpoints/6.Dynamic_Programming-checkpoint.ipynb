```python
import graphviz
import os
```

```python
# N:ナップサックの個数 
# W:持てる重さ
# w: 重量
# v: 価値

N = 6
W = 9
w = [2, 1, 3, 2, 1, 5]
v = [3, 2, 6, 1, 3, 10]

dp = [[0]*(W+1) for i in range(N+1)]
```

```python
G = graphviz.Digraph(engine='neato', format='svg', strict=True)

def print_graph(G):
    for i in range(N+1):
        for g in range(W+1):
            G.node(str(i)+str(g), label=str(dp[i][g]), pos=str(g*0.6)+","+str((N-i)*0.6)+"!", shape="circle", fixedsize="true")
    for i in range(N+1):
        G.node(str(i)+" item", pos=str(-0.6)+","+str((N-i)*0.6)+"!", style="filled", color="#00000000")
    for g in range(W+1):
        G.node(str(g)+" g", pos=str(g*0.6)+","+str((N+0.9)*0.6)+"!", style="filled", color="#00000000")
    display(G)
    
picnum = 0
def save_graph(name=""):
    global picnum
    if name == "":
        name = str(picnum)
        picnum += 1
    G.render("svg/"+name)
    os.remove("svg/"+name)
```

```python
def add_arrow(s1, s2, g1, g2, color="black"):
    G.edge(str(s1)+str(s2), str(g1)+str(g2), color=color)

def add_color(s1, s2, g1, g2):
    G.edge(str(s1)+str(s2), str(g1)+str(g2), color="red")
    #G.node(str(s1)+str(s2), style="filled", fillcolor="yellow")
    G.node(str(g1)+str(g2), style="filled", fillcolor="yellow")
```

```python
print_graph(G)
#save_graph("default")
```

```python
for i in range(N):
    for g in range(W+1):
        if g >= w[i] and dp[i][g-w[i]]+v[i] > dp[i][g]:
            dp[i+1][g] = dp[i][g-w[i]]+v[i]
            add_arrow(i, g-w[i], i+1, g)
            continue
        dp[i+1][g] = dp[i][g]
        add_arrow(i, g, i+1, g, "gray")
        
    print_graph(G)
    #save_graph()

print(dp[N][W])
```

```python
add_color(2,0,3,3)
add_color(4,3,5,4)
add_color(5,4,6,9)
display(G)
#save_graph("final")
```
