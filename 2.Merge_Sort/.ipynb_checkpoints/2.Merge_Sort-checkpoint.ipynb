```python
from collections import deque
import graphviz
import os
```

```python
#N: 入力数
#A: 配列
N = 4
A = [4, 3, 2, 1]
```

```python
def print_graph(A):
    for i, a in enumerate(A):
        G.node(str(a), shape="circle", pos=str(i*0.6)+",0!")
    display(G)

picnum = 0
def save_graph(name=""):
    global picnum
    if name == "":
        name = str(picnum)
        picnum += 1
    G.render("svg/"+name)
    os.remove("svg/"+name)
```

```python
G = graphviz.Graph(engine='neato', format='svg', strict=True)

print_graph(A)
#save_graph("default")
```

```python
def merge(A, left, mid, right):
    n1 = mid - left
    n2 = right - mid
    L = [0]*(n1+1)
    R = [0]*(n2+1)
    for i in range(n1):
        L[i] = A[left+i]
    for i in range(n2):
        R[i] = A[mid+i]
    L[n1] = float("inf")
    R[n2] = float("inf")
    i = 0
    j = 0
    for k in range(left, right):
        if L[i] <= R[j]:
            A[k] = L[i]
            i += 1
        else:
            A[k] = R[j]
            j += 1

def mergeSort(A, left, right):
    if right - left > 1:
        mid = (left+right) // 2
        mergeSort(A, left, mid)
        mergeSort(A, mid, right)
        merge(A, left, mid, right)
        print_graph(A)
        #save_graph()


mergeSort(A, 0, N)
```
