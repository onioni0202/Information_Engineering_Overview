```python
from collections import defaultdict
import networkx as nx
from IPython.display import SVG, display
import pydot
import os
```

```python
#N:ノード数 M:エッジの数
#xy: エッジの始点と終点

N = 8
M = 7
xy = [(1, 2),
      (2, 4),
      (2, 5),
      (8, 6),
      (6, 7),
      (1, 3),
      (3, 8)]

d = defaultdict(list)

for i in range(M):
    d[xy[i][0]].append(xy[i][1])
    d[xy[i][1]].append(xy[i][0])
```

```python
G = nx.Graph()
G.add_edges_from(xy)
graph = nx.nx_agraph.to_agraph(G)
graph.node_attr["shape"] = "circle"
graph.node_attr["style"] = "filled"

def print_graph(nodes):    
    for i, b in enumerate(nodes, 1):
        node = graph.get_node(i)
        if b:
            node.attr["fillcolor"] = "red"
        else:
            node.attr["fillcolor"] = "white"
    
    svg = SVG(graph.draw(prog='dot', format='svg'))
    display(svg)

picnum = 0
def save_graph(name=""):
    global picnum
    if name == "":
        name = str(picnum)
        picnum += 1
    graph.write("svg/"+name+".dot")
    (f,) = pydot.graph_from_dot_file("svg/"+name+".dot")
    f.write_svg("svg/"+name+'.svg')
    os.remove("svg/"+name+".dot")
```

```python
nodes = [False]*N
stack = []
print_graph(nodes)
#save_graph("default")

s = 1 # スタート
stack.append(s)
nodes[s-1] = True

while len(stack)!=0:
    From = stack.pop()
    for To in d[From]:
        if nodes[To-1] == False:
            print_graph(nodes)
            #save_graph()
            stack.append(To)
            nodes[To-1] = True
print_graph(nodes)
#save_graph()
```
